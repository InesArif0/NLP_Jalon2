```python
def display_topics(model, feature_names, num_top_words, dftopic,df, topic_names=None):

    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            reviews =[]
            print("\nTopic ", ix)
           
            reviews = [ df['text'][i] for i in dftopic.nlargest(3,ix).index ]
            print("\nreview1 : ", reviews[0])
            print("\nreview2 : ",reviews[1])
            print("\nreview3 : ",reviews[2])
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i] \
             for i in topic.argsort()[:-num_top_words - 1:-1]]))
```

```python
nmf_model = NMF(15)

doc_topic = nmf_model.fit_transform(dataset_tfid)
```

```python
dftopic = pd.DataFrame(doc_topic)
dftopic
```

```python
display_topics(nmf_model, feature_names, 6,  dftopic, df_1_2)
```
